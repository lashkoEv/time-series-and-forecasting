#### Mean method

```python
df_mean = df["value"].mean()

print("Mean: ", df_mean)

forecast_dates = pd.date_range(start=df['date'].iloc[-1], periods=25, freq='MS')[1:]
forecast_values = [df_mean] * len(forecast_dates)
forecast_df_mean = pd.DataFrame({'date': forecast_dates, 'forecast_value': forecast_values})

display(forecast_df_mean)

plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['value'], marker='o', label='Data')
plt.axhline(y=df_mean, color='g', linestyle='--', label='Mean')
plt.plot(forecast_df_mean['date'], forecast_df_mean['forecast_value'], marker='o', linestyle='--', color='r', label='Forecast data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

#### Naive method 


```python
df_last = df["value"].iloc[-1]

print("Last value: ", df_last)

forecast_dates = pd.date_range(start=df['date'].iloc[-1], periods=25, freq='MS')[1:]
forecast_values = [df_last] * len(forecast_dates)
forecast_df_naive = pd.DataFrame({'date': forecast_dates, 'forecast_value': forecast_values})

display(forecast_df_naive)

plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['value'], marker='o', label='Data')
plt.plot(forecast_df_naive['date'], forecast_df_naive['forecast_value'], marker='o', linestyle='--', color='r', label='Forecast data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

#### Seasonal naive method

```python
df_last_season = df["value"].tail(12)

print("Last season values: ")
display(df_last_season)

forecast_dates = pd.date_range(start=df['date'].iloc[-1], periods=25, freq='MS')[1:]
forecast_values =  pd.concat([df_last_season, df_last_season], ignore_index=True)
forecast_df_season = pd.DataFrame({'date': forecast_dates, 'forecast_value': forecast_values})

display(forecast_df_season)

plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['value'], marker='o', label='Data')
plt.plot(forecast_df_season['date'], forecast_df_season['forecast_value'], marker='o', linestyle='--', color='r', label='Forecast data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## 3. Для кожного з наївних методів оцініть точність прогнозування на 1 крок уперед у такий спосіб:
#### - Розбийте ряд на 2 частини: навчальну та тестову. 
#### - В навчальну помістіть 70-80% перших точок ряду, інші точи утворять тестову частину. 
#### - Використовуючи навчальну частину, зробіть прогноз на 1 крок уперед і розрахуйте абсолютну помилку прогнозування, 
#### порівнявши прогнозне значення зі справжнім значенням з тестової частини.
#### - Додайте до навчальної частини 1 точку з тестової, знову зробіть проноз і розрахуйте помилку. 
#### - Робіть так, поки не завершиться тестова частина. 
#### - Тоді на основі отриманого масиву помилок розрахуйте метрики MAE та RMSE. 
#### - Ці метрики мають бути розраховані для кожного методу прогнозування. 
#### - Який з методів дає точніший прогноз згідно розрахованих метрик?

```python
def get_metrics_mean(train, test) :
    predictions = []

    for i in range(len(test)):
        forecast = train['value'].mean()
        predictions.append(forecast)
        train = pd.concat([train, test.iloc[[i]]])

    mae = mean_absolute_error(test['value'], predictions)
    rmse = np.sqrt(mean_squared_error(test['value'], predictions))
    
    return mae, rmse, predictions

def get_metrics_naive(train, test) :
    predictions = []

    for i in range(len(test)):
        forecast = train['value'].iloc[-1]
        predictions.append(forecast)
        train = pd.concat([train, test.iloc[[i]]])

    mae = mean_absolute_error(test['value'], predictions)
    rmse = np.sqrt(mean_squared_error(test['value'], predictions))
    
    return mae, rmse, predictions

def get_metrics_seasonal(train, test) : 
    predictions = []

    for i in range(len(test)):
        season = 12
        forecast = train['value'].iloc[-season]
        predictions.append(forecast)
        train = pd.concat([train, test.iloc[[i]]])

    mae = mean_absolute_error(test['value'], predictions)
    rmse = np.sqrt(mean_squared_error(test['value'], predictions))
    
    return mae, rmse, predictions
```

```python
train_size = int(len(df) * 0.7)
train, test = df[:train_size], df[train_size:]

mae_mean, rmse_mean, data_mean = get_metrics_mean(train, test)
mae_naive, rmse_naive, data_naive = get_metrics_naive(train, test)
mae_seasonal, rmse_seasonal, data_seasonal = get_metrics_seasonal(train, test)

table = pd.DataFrame(columns={"Method" : 1, "MAE": 2, "RMSE": 3})
table.loc[len(table)] = ["Mean", mae_mean, rmse_mean]
table.loc[len(table)] = ["Naive", mae_naive, rmse_naive]
table.loc[len(table)] = ["Seasonal", mae_seasonal, rmse_seasonal]
display(table.style.hide_index()) 

df_mean = pd.DataFrame({'date': test['date'], 'value': data_mean})
df_naive = pd.DataFrame({'date': test['date'], 'value': data_naive})
df_seasonal = pd.DataFrame({'date': test['date'], 'value': data_seasonal})

plt.figure(figsize=(30, 5))
plt.plot(df['date'], df['value'], marker='o', label='Data')
plt.plot(df_mean['date'], df_mean['value'], marker='o', label='Mean method')
plt.plot(df_naive['date'], df_naive['value'], marker='o', label='Naive method')
plt.plot(df_seasonal['date'], df_seasonal['value'], marker='o', label='Seasonal method')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

#### Через яскраво виражену сезонність даних очевидним є висновок, що найкращі результати дає сезонний наївний метод
